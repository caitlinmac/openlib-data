```python
import requests
import pandas as pd
```

### Return first page of API query call

```python
""" 
API CALL: Return 1st page
"""
"https://openlibrary.org/search.json?q=first_publish_year:2019&fields=first_publish_year,title,author_name,language,number_of_pages_median,publisher,format,subject,ratings_average,ratings_sortable,ratings_count,readinglog_count"

# endpoint URL for Open Library search API
url = "https://openlibrary.org/search.json"

# Query parameters
q_params = {
    'q': 'first_publish_year:2019 AND (NOT publisher:Independently Published)',
    'sort': 'random',
    'fields': 'first_publish_year,title,author_name,language,number_of_pages_median,publisher,format,subject,ratings_average,ratings_sortable,ratings_count,readinglog_count',
    'page': 1,  # Start with the first page
    'limit': 100000  # Number of results per page
}
```

```python
%%time
api_data = requests.get(url, params=q_params).json()
```

```python
len(api_data['docs'])
```
